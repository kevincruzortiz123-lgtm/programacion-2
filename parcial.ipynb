```python
A={1,2,3,4,5}
B={5,6,7,8,9}
```

```python
A.difference(B)
```

el metodo .difference() quita del conjunto A los elementos que comparte con el conjunto B que este caso era el 5 al compartirlo ambos conjuntos A y B. el conjunto resultante seria el que tiene los elementos {1,2,3,4}

```python
A.symmetric_difference(B)
```

el metodo .symmetric_difference() da los elementos que no comparten entre ellos siendo en este caso el 5 el que comparten y por eso el unico que no aparece. el conjunto resultante seria el que tiene los elementos{1,2,3,4,6,7,8,9}

```python
A={1,2,3,5,6,7,8,9}
B={5,6,7}
```

```python
A.issuperset(B)
```

el metodo .issuperset() verifica que todos los elementos de un conjunto se encuentran dentro de otro conjunto. El valor que devuelve es true ya que todos valores de B estan dentro de A

```python
A.isdisjoint(B)
```

el metodo .isdisjoint() verifica que ningun elemento de un conjunto se encuentre dentro de otro. Devuelve false si al menos un elemento se encuentra contenido en otro conjunto

```python
def m(m):
  print("m=",m)
  if isinstance(m, str):
    print("tiene que ser un numero")
  elif isinstance(m, int) and m >= 0:
    for n in range(0, (m+1)):
      if n == 1 or n == 0:
        print("El número es ", n)
      else:
        for x in range(2, n):
          if n % x == 0:
            print(n, 'es igual a', x, '*', n // x)
            break
        else:
          print(n, 'es un número primo')
  elif m < 0:
    print("El número no es natural")
  elif isinstance(m, float):
    print("es un numero decimal")

```

```python
m("a")
print("")
m(3.8)
print("")
m(-3)
print("")
m(5)
print("")
```
