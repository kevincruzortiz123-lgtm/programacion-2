```python
np.random.seed(0)
prueba_A= np.random.normal(loc=7.2, scale=0.4,size=52)
prueba_B= np.random.gamma(shape=306, scale=0.0229, size=52)
examen=pd.DataFrame({
    "prueba_A":prueba_A,
    "prueba_B":prueba_B
})
```

```python
examen.max()
```

```python
examen.min()
```

```python
examen.describe()
```

```python
from scipy import stats
```

```python
from scipy.stats import shapiro
shapiro(examen["prueba_A"],)
```

como el valor p es mayor que 0.05, no se rechaza la hipotesis de normalidad con una confianza del 95%, entonces se cumple e supuesto de normalidad

```python
shapiro(examen["prueba_B"])
```

como el valor p es mayor que 0.05, no se rechaza la hipotesis de normalidad con una confianza del 95%, entonces se cumple e supuesto de normalidad

```python
t_stat, p_val = stats.ttest_rel(examen["prueba_A"], examen["prueba_B"])
print("Estadístico t:", t_stat)
print("Valor p:", p_val)
```

No se rechaza que el puntaje medio en ambas encuestas es igual con una confianza del 95%
