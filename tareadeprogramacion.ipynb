```python
lista=['carro', 'moto', 'bus', 'avion']
print(lista)
tupla=tuple(lista)
print(tupla)
diccionario=dict(enumerate(tupla))
print(diccionario)
conjunto=set(diccionario.items())
print(conjunto)
```

```python
tupla=('luna', 'ana', 'sergio', 'pedro')
print(tupla)
conjunto=set(tupla)
print(conjunto)
lista=list(conjunto)
print(lista)
diccionario=dict(enumerate(lista))
print(diccionario)
```

```python
diccionario={'nombre': 'juan','edad': 25, 'ocupacion': 'profesor de la UIS'}
print(diccionario)
lista=list(diccionario.values())
print(lista)
conjunto=set(lista)
print(conjunto)
tupla=tuple(conjunto)
print(tupla)
```

```python
conjunto={'taller de lenguaje', 'algebra lineal I', 'calculo I', 'geometria euclidiana', 'programacion'}
print(conjunto)
diccionario=dict(enumerate(conjunto))
print(diccionario)
tupla=(tuple(diccionario.values()))
print(tupla)
lista=list(tupla)
print(lista)
```
