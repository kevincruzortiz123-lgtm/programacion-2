```python
%%writefile menu.py
def mostrar_menu():
  print("MENU DE OPERACIONES CON MATRICES")
  print("1. Suma de matrices")
  print("2. Multiplicacion de matrices")
  print("3. Producto de hadamard")
  print("4. Producto de kronecker")
  print("5. Salir")

  while True:
    try:
      opcion = int(input("Ingrese una opcion: "))
      if opcion in [1,2,3,4,5]:
        return opcion
      else:
        print("Opcion invalida. Intente de nuevo.")
    except ValueError:
      print("Error. Ingrese un numero valido.")
```

```python
%%writefile entrada.py
def ingresar_matrices():
    print("PRIMERA MATRIZ")
    filas = int(input("Ingrese el número de filas: "))
    columnas = int(input("Ingrese el número de columnas: "))
    matriz = []
    for i in range(filas):
        while True:
            fila = list(map(float, input(f"Ingrese la fila {i+1}: ").split()))
            if len(fila) != columnas:
                print(f"Error: debe ingresar {columnas} valores.")
            else:
                matriz.append(fila)
                break
    print("SEGUNDA MATRIZ")
    filas2 = int(input("Ingrese el número de filas: "))
    columnas2 = int(input("Ingrese el número de columnas: "))
    matriz2 = []
    for i in range(filas2):
        while True:
            fila2 = list(map(float, input(f"Ingrese la fila {i+1}: ").split()))
            if len(fila2) != columnas2:
                print(f"Error: debe ingresar {columnas2} valores.")
            else:
                matriz2.append(fila2)
                break
    return matriz, filas, columnas, matriz2, filas2, columnas2
def mostrar_matrices(matriz, matriz2):
    print("PRIMERA MATRIZ")
    for fila in matriz:
        print(fila)
    print("SEGUNDA MATRIZ")
    for fila in matriz2:
        print(fila)
```

```python
%%writefile operaciones_matices.py
def suma_matrices(matriz, matriz2):
  if filas != filas2 or columnas != columnas2:
   print("Error: Las matrices deben tener las mismas dimensiones para sumarse.")
  else:
      resultado = []
      for i in range(filas):
            fila = []
            for j in range(columnas):
              fila.append(matriz[i][j] + matriz2[i][j])
            resultado.append(fila)
      print("la suma es:",resultado)
def Multiplicacion_matrices(matriz, matriz2):
        if columnas != filas2:
           print( "Error: El número de columnas de A debe ser igual al número de filas de B.")
        resultado = []
        for i in range(filas):
          fila = []
          for j in range(columnas2):
            suma = 0
            for k in range(columnas):
              suma += matriz[i][k] * matriz2[k][j]
            fila.append(suma)
          resultado.append(fila)
        print("el producto es: ", resultado)
def producto_hadamard(matriz, matriz2):
        if filas != filas2 or columnas != columnas2:
          print("Error: Las matrices deben tener las mismas.")
        else:
          resultado = []
          for i in range(filas):
            fila = []
            for j in range(columnas):
              fila.append(matriz[i][j] * matriz2[i][j])
            resultado.append(fila)
          print("el producto de hadamart es:",resultado)
def producto_kronecker(matriz, matriz2):
        resultado = []
        for i in range(filas):
          for k in range(filas2):
           fila = []
           for j in range(columnas):
              for l in range(columnas2):
                fila.append(matriz[i][j] * matriz2[k][l])
           resultado.append(fila)
        print("el producto de Kronecker es: ", resultado)
```
