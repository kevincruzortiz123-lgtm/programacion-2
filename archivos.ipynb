```python
with open("archivo.txt","w")as f: #con "w" crea un archivo
  f.write("hola, este es un archivo de texto.\n")
  f.write("segunda línea.\n")
```

```python
with open("archivo.txt","r")as f: #con "r" se lee el archivo
  print(f.read())
```

```python
import csv

ruta = "/content/drive/MyDrive/datos.csv"

# Leer
with open(ruta, "r") as f:
    data = list(csv.reader(f))

# Mostrar antes
print("ANTES:")
print(data)

# Agregar fila
data.append(["Diego", "28", "Armenia"])

# Guardar
with open(ruta, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

# Mostrar después
print("\nDESPUÉS:")
with open(ruta, "r") as f:
    print(f.read())
```
