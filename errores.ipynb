```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()
    password = input("Contraseña: ")

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    if usuarios[usuario] != password:
        raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)
    quit()

except ValueError as e:
    print("Error:", e)
    quit()
```

```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    elif usuarios[usuario] != input("Contraseña: "):
      raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

la diferencia entre ambos codigos es que el primero pide usuario y contraseña antes de verificar si el usuario y contraseña son correctos, el segundo primero verifica que el ususrio exista para luego pedir la contraseña

```python
try:
  num1= float(input("numero 1: "))
  num2= float(input("numero 2: "))
  op = input("operacion (+,-,*,/,%): ")

  if op == "+":
    resultado = num1 + num2
  elif op == "-":
    resultado = num1 - num2
  elif op == "*":
    resultado = num1 * num2
  elif op == "/":
    resultado = num1 / num2
  elif op=="%":
    resultado= num1 % num2
  else:
    raise ValueError("operacion no valida")

  print("resultado:", resultado)

except ValueError as e:
  print("error de valor: ", e)
  quit()
except TypeError as e:
  print("error de tipo: ", e)
  quit()
except ZeroDivisionError as e:
  print("error de division: ", e)
  quit()
```

```python
try:

    datos = input("Ingresa notas separadas por coma: ")

    notas = [float(n) for n in datos.split(",")]

    if len(notas) == 0:
        raise ValueError("No ingresaste notas")

    promedio = sum(notas) / len(notas)

    posicion = int(input("¿Qué nota quieres ver (índice)?: "))

    if posicion < 0:
        raise ValueError("Ingresa solo índices positivos")

    print("Nota seleccionada:", notas[posicion])

    print("Promedio:", promedio)

except ValueError as e:
    print("Error de valor:", e)

except ZeroDivisionError:
    print("No se puede dividir entre cero")

except IndexError:
    print("Índice fuera de rango")
```
