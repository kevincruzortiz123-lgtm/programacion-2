```python
%%writefile main.py
import menu
import entrada
import operaciones_matices
while True:
 opcion = menu.mostrar_menu()
 if opcion == 5:
    print("Gracias por usar el programa")
    break
 matriz, filas, columnas, matriz2, filas2, columnas2 = entrada.ingresar_matrices()
 entrada.mostrar_matrices(matriz, matriz2)
 operaciones_matices.filas = filas
 operaciones_matices.columnas = columnas
 operaciones_matices.filas2 = filas2
 operaciones_matices.columnas2 = columnas2

 if opcion == 1:
    operaciones_matices.suma_matrices(matriz, matriz2)
 elif opcion == 2:
    operaciones_matices.Multiplicacion_matrices(matriz, matriz2)
 elif opcion == 3:
    operaciones_matices.producto_hadamard(matriz, matriz2)
 elif opcion == 4:
    operaciones_matices.producto_kronecker(matriz, matriz2)
```
